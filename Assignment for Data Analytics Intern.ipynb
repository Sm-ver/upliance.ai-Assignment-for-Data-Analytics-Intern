# 4. Merging Data  

I am joining these given table, firstly checking common columns betweem cooking session and order details table and joining them
then, I am joinin the resultant table with user details table to get final table.

```python
print(set(df_cooking_session.columns)&set(df_order_details.columns))
```

```python
df_join1 = pd.merge(df_cooking_session,df_order_details,on=['Meal Type', 'Dish Name', 'Session ID', 'User ID'])
df_join1
```

```python
print(set(df_join1.columns)&set(df_user_details.columns))
```

```python
# joining resultant from 1st join with user details table, to get final table
df_final = pd.merge(df_join1,df_user_details,on=['User ID'],how='left')
df_final
```

# 6. Insights and Plots

```python
plt.figure(figsize=(10,4))

x_axis = 'Dish Name'
y_axis = 'Duration (mins)'

sns.barplot(x=x_axis, y=y_axis, data=df_final)

plt.title('1. Duration (mins) for Different Dishes')
plt.xlabel(x_axis)
plt.ylabel(y_axis)
plt.show()
```

1. From the above plot we can see, Grilled Chicken takes maximum time to cook, while Oatmeal takes least time making it suitable for breakfast.

```python
plt.figure(figsize=(10,4))

x_axis = 'Dish Name'
y_axis = 'Session Rating'

sns.barplot(x=x_axis, y=y_axis, data=df_final)

plt.title('2. Session Rating for Different Dishes')
plt.xlabel(x_axis)
plt.ylabel(y_axis)
plt.show()
```

2. From the above plot we can see, Grilled Chicken is having highest session rating while Oatmeal is having lowest.(Assuming "Session Rating" is rating for the Session and "Ratings" columns is ratings for the Order) 

```python
plt.figure(figsize=(10,4))

x_axis = 'Meal Type'
y_axis = 'Duration (mins)'

sns.barplot(x=x_axis, y=y_axis, data=df_final)

plt.title('3. Duration (mins) for Different Meal Types')
plt.xlabel(x_axis)
plt.ylabel(y_axis)
plt.show()
```

3. From above plot we can see, people spend the most time on dinner, while lunch and breakfast are quicker meals. Breakfast times vary the most between different people.

```python


plt.figure(figsize=(10,4))

x_axis = 'Dish Name'
y_axis = 'Rating'

print(df_final[df_final['Order Status']=='Completed'].groupby([x_axis])[y_axis].mean())


sns.barplot(x=x_axis, y=y_axis, data=df_final[df_final['Order Status']=='Completed'])

plt.title('4. Order Rating for Different Dishes')
plt.xlabel(x_axis)
plt.ylabel(y_axis)
plt.show()
```

4. From the above plot we can see, Grilled Chicken and Spaghetti have highest ratings while order dishes and lower ratings in similar range . Highest rating is of Grilled Chicken having 4.67 rating.

```python
df_final['Age Group'] = pd.cut(df_final['Age'], bins=[20, 30, 40, 50, 60], labels=['20-29', '30-39', '40-49', '50-59'], right=False)

# Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=df_final, x='Age Group', hue='Dish Name')

# Labels and title
plt.title('12 .Favorite Dish by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count of Dish')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

12. From above plot we can see, for younger people (below 40) Spaghetti and Caesar Salad is popular, Grilled Chicken is popular among all age groups. Oatmeal is only popular among 30-39 age group.

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=df_final, x='Location', hue='Dish Name')


plt.title('13 .Favorite Dish by Cities')
plt.xlabel('Age Group')
plt.ylabel('City')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
